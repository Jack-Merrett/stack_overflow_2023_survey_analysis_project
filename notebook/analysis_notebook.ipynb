## Analysis of Data


### 1. What jobs paid the most at?

In order to do this we are going to find the Average amount on money people get paid, grouped by the "DevType". 


```python
# We must use a filtered df to not skew the numbers.
filt_df = df[df['DevType'] != "Unknown"] # filtering out unknown jobs
filtered_df = filt_df[filt_df['ConvertedCompYearly'] != 1] # filtering out 1s
```

```python
# Group by 'DevType' and calculate the average salary for each group
average_salary_by_devtype = filtered_df.groupby('DevType')['ConvertedCompYearly'].mean().sort_values(ascending = False).reset_index()
```

```python
# Loop through the groups and print the average salary
counter = 0
print('The top 5 most high-playing by dev type:')
for index, row in average_salary_by_devtype.iterrows():
    devtype = row['DevType']
    avg_salary = row['ConvertedCompYearly']
    if counter >= 5: # we only want the top 5 jobs
        break
    print(f"Job Title: {devtype}, Average Yearly Salary: {avg_salary:.2f}")
    counter += 1
```

#### Conclution: 
Looks like an **Experienced Developer** makes the most money, with an average  yearly salary of: **$301,328.45**
